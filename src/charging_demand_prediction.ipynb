```python
# Load the travel demand data - add path to the "GDOT_2019_09.csv"
data = pd.read_csv('') 
```

```python
data.head()
```

```python
# Load the estimated EV registration data in GA - path to "georgia_place_infrastructure_emb_zip_code_avg_autoenc_64_predicted.csv"
ga = pd.read_csv('')
```

```python
# data is in bin number, need to convert to real vehicle count
import numpy as np
bin_ranges = {
    0: (1, 13),
    1: (13, 52),  # Bin 1 has range from 0 to 10
    2: (52, 229),  # Bin 2 has range from 10 to 20
    3: (229, 3287)   # Bin 3 has range from 20 to 30
    
}
def get_random_within_bin(bin_num):
    min_val, max_val = bin_ranges[bin_num]
    return np.random.uniform(min_val, max_val)

# Apply the function to generate a random number within the bin range
ga['ev_number'] = ga['predicted_ev_count_bin'].apply(get_random_within_bin)
```

```python
# Normalized with actual GA ev number
ga['ev_number'] = np.ceil(ga['ev_number']/sum(ga['ev_number']) * 92400)
```

```python
ga.head(10)
```

```python
# Selected trips within GA
trip = data.loc[data['origin_state'] == 'GA',:]
trip = trip.loc[trip['destination_state'] == 'GA',:]
```

```python
# Selected trips of weekday
trip_wd = trip.loc[trip['day_type'] == 'Weekday']
trip_we = trip.loc[trip['day_type'] == 'Weekend']
```

```python
# Calculate the trip distribution per origin
trip_wd['count'] = trip_wd.groupby('origin_zone_id')['total_linked_trips'].transform('sum')
trip_wd['distribution'] = trip_wd['total_linked_trips']/trip_wd['count']
```

```python
trip_wd['distribution'].describe()
```

```python
# Adjust the zone id to int
trip_wd['origin_zone_id'] = trip_wd['origin_zone_id'].astype(int)
```

```python
# Match the trip table with EV registration
df = trip_wd.merge(ga.loc[:,['geohash','ev_number']], left_on = 'origin_zone_id', right_on = 'geohash', how = 'left')
```

```python
# Selected top 5 destinations for each origin
test = df.sort_values(['origin_zone_id','distribution'], ascending = False).groupby('origin_zone_id').head(5)
```

```python
# Normalized the distribution among selected destinations
test['total'] = test.groupby('origin_zone_id')['distribution'].transform('sum')
test['new_distribution'] = test['distribution'] / test['total']
test['ev_number'] = test['ev_number'].fillna(0)
```

```python
# Estimated number of EV trips per destination

test['ev_trips'] = np.ceil(test['ev_number'] * test['new_distribution'])
```

```python
test['ev_trips'].sum()
```

```python
# Estimated the EV charging demand by the maximum of EV trips or EV registration per census block, considering
# both the demand of home charging and en-route charging
test['destination_zone_id'] = test['destination_zone_id'].astype(int)
demand = test.groupby('destination_zone_id')['ev_trips'].sum()
demand = demand.reset_index()
demand.columns = ['id', 'count']
total_demand = ga.merge(demand, left_on = 'geohash', right_on = 'id', how = 'left')
total_demand['geohash'] = total_demand['geohash'].fillna(total_demand['id'])
total_demand['ev_number'] = total_demand['ev_number'].fillna(0)
total_demand['count'] = total_demand['count'].fillna(0)
total_demand['demand'] = total_demand[['ev_number', 'count']].max(axis = 1)
total_demand['geohash'] = total_demand['geohash'].astype(int)
```

```python
# Load the existing EV stations "Existing ev charging stations.csv"
ev_stations = pd.read_csv('')
# ev_stations['join_id'] = ev_stations['join_id'].astype(int)

# Count the number of existing chargers per census block
chargers = ev_stations.groupby('join_id')['chargers'].sum()
chargers = chargers.reset_index()

# Match the existing charger number with the demand table
demand = total_demand.merge(chargers, left_on = 'geohash', right_on = 'join_id', how = 'left')
```

```python
# Output the final charging demand file - "final_demand.csv"
demand['join_id'] = demand['join_id'].fillna(demand['geohash'])
demand['chargers'] = demand['chargers'].fillna(0)
demand.loc[:,['geohash', 'ev_number', 'count', 'demand','chargers']].to_csv("",index = None)
```
